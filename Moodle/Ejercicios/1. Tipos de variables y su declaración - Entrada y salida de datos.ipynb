## Ejercicio 3 - Economista 2

Como su predicción en el ejercicio anterior fue correcta, ahora le contrataron para predecir el valor de muchas más acciones dentro de muchas semanas.

Usted sabe que el valor de las acciones dentro de una cantidad de semanas está determinado por:

1. Semanas de espera (int): tiempo en el que determinará el valor que tomará la acción.
2. Porcentaje de subida (float): valor positivo que determina la subida del precio de la acción.
3. Porcentaje de bajada (float): valor negativo que determina la bajada del precio de la acción.

La fórmula matemática es la siguiente:

$$
1 + \frac{Porcentaje\ de\ subida + Porcentaje\ de\ bajada}{Semanas\ de\ espera}
$$

Al ser tantas predicciones las que le piden hacer, usted decidió crear en python un programa que automatice dicha labor.
El programa en Python debe imprimir el resultado de la fórmula.

Ejemplo:

    Entrada:
        Semanas de espera = 5
        Porcentaje de subida = 0.5
        Porcentaje de bajada = -1.12
        
    Procedimiento:
$$
1 + \frac{0.5 + (-1.12)}{5}
$$

    Salida: 0.876



<div class="alert alert-success">

`Solución economista2.py`
```python
semanas = int(input())
subida = float(input())
bajada = float(input())

numerador = subida + bajada
denominador = semanas
print(1 + (numerador/denominador))
```

## Ejercicio 4 - Concierto de Piano

Usted toca el piano muy bien, y le invitaron al auditorio fundadores a dar un concierto. Debido a que su trabajo cuesta, para ingresar se debe pagar una boleta, sin embargo, usted solo se queda con una parte del pago de la boleta porque la universidad debe pagar la luz, las instalaciones, etc.

Para calcular la cantidad de dinero con la que se queda usted necesita saber los siguientes datos:
- Cantidad de personas planta baja (int)
- Precio boleta planta baja (float)
- Cantidad de personas planta alta (int)
- Precio boleta planta alta (float)

La operación matemática que determina la cantidad de dinero que usted se queda al final es la siguiente:

$$
\frac{Personas\ planta\ baja \cdot Precio\ planta\ baja }{2} + {Personas\ planta\ alta \cdot Precio\ planta\ alta^2}
$$

Realice un programa en Python que resuelva esa ecuación e imprima el resultado.

__Importante:__
- Para la potencia haga lo siguiente:

    número_elevado_2 = número**2
    
    el operador '^' en Python __no__ es usado para la potencia, así que le podría dar error en el resultado.
    
    
__Nota:__ Elabore su programa en un archivo de Python llamado **conciertoDePiano.py**.

<div class="alert alert-success">

`Solución conciertoDePiano.py`
```python
personas_planta_baja = int(input())
precio_planta_baja = float(input())
personas_planta_alta = int(input())
precio_planta_alta = float(input())

parte1 = (personas_planta_baja * precio_planta_baja)/2
parte2 = personas_planta_alta * (precio_planta_alta**2)

print(parte1+parte2)
```
